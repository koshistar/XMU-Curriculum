#### 例1-1 读入excel文件“WPP2022_人口统计学数据.xlsx”的数据，
    （1）只显示LocTypeName、Local、Time、To1July、PopDensity、PopSexRatio、PopGrowthRate、DoublingTime、TFR
    （2）设置"LocTypeName"为行索引

```python
import pandas as pd
pd.set_option('display.float_format', lambda x: '%.1f' % x)  #设置全局显示选项，避免科学记数法，保留一位小数
df1=pd.read_excel(r'.\WPP2022_人口统计学数据.xlsx',usecols=["LocTypeName","Local","Time","To1July","PopDensity","PopSexRatio","PopGrowthRate","DoublingTime","TFR"])
df1.set_index("LocTypeName",inplace=True)
df1
```

## 一、数据查询

```python
#练习1，查询第150-160条记录的情况
df1[150:160]

```

```python
#练习2，查询第150-160条记录"Local","Time","To1July"列的情况
df1[150:160][["Local","Time","To1July"]]

```

```python
#练习3，查询中国各年(Time)的人口密度PopDensity、总生育率TFR。(Local是China)
#结果保存在cx3中
cx3=df1[(df1.Local=='China')][['Time','PopDensity','TFR']]
cx3
```

```python
#练习4，设置cx3以Time为索引，原地修改，用describe查询cx3的统计数据。
cx3.set_index('Time',inplace=True)
cx3.describe()
```

```python
#练习5，查询中国各年(cx3中）的人口密度PopDensity在[114.6,119.2]之间的年份及总生育率
# cx3[(cx3['PopDensity']>=114.6) & (cx3['PopDensity']<=119.2)][['PopDensity','TFR']]
cx3.loc[(cx3['PopDensity']>=114.6) & (cx3['PopDensity']<=119.2),['PopDensity','TFR']]


```

```python
#练习6，查询cx3，中国人口密度PopDensity>139 并且 总生育率TFR<1.3的年份，  
#注意：& 运算符 两边的条件必须加()，也即是 ( 条件1 )  & ( 条件2  )
# cx3[(cx3.PopDensity>139) & (cx3.TFR<1.3)][['PopDensity','TFR']]
cx3.loc[(cx3.PopDensity>139) & (cx3.TFR<1.3),['PopDensity','TFR']]

```

```python
#练习7，查询df1，选出其中China与Africa的2024年的数据
# df1[(df1['Time']==2024) & (df1['Local'].isin(['China', 'Africa']))]
df1.loc[(df1['Time']==2024) & (df1['Local'].isin(['China', 'Africa']))]



```

# 二、排序
1. 按行或列索引排序：    sort_index(axis=0, ascending=True, inplace=False) 
2. 按值排序：    DataFrame.sort_values(by, axis=0,ascending=True, inplace=False, na_position='last')

```python
#练习1，将df1中的数据按列标签升序排序
df1.sort_index(axis=1, ascending=True,inplace=False)

```

```python
#练习2 将df1中的数据按行标签降序排序
df1.sort_index(axis=0,ascending=False)

```

```python
#练习3 将df1中的数据按人口增长率PopGrowthRate的降序排序,只显示最多的8个
df1.sort_values(by='PopGrowthRate',ascending=False).head(8)

```

```python
df2=df1.copy()
#练习4，将df2中的数据先按“地区名”Local的升序,
#地区名相同的则按人口增长率PopGrowthRate的降序排序，原地排序
df2.sort_values(by=['Local','PopGrowthRate'],ascending=[True,False],inplace=True)
df2
```
