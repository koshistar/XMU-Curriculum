# 一、处理缺失值
###  (1)查询缺失值
#### 获取列与特定值匹配的行的索引:
    info与isnull方法    
###  (2)删除缺失值
    DataFrame.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False) 
###  (3)填充缺失值
    DataFrame.fillna(value, axis=0, method, limit,inplace=False)       
         说明 ：method='ffill'表示前向填充， method='bfill'表示后向填充，以后的版本可能不支持method参数
    DataFrame.ffill(value, axis=0, limit,inplace=False)     #前一行的值来填充当前行的缺失值。
    DataFrame.bfill(value, axis=0, limit,inplace=False)    #后一行的值来填充当前行的缺失值。

#### 例1-3 填充值

```python
df_f1=df.copy()

#查看 商品类别 为空值的记录
df_f1[df_f1["商品类别"].isnull()]
```

```python
#求 商品类别 的众数，用众数填充 商品类别 空值
#求 单价的均值，用均值填充 单价的空值
a=df_f1['商品类别'].mode()
print("商品类别的众数是：\n",a)                                          #注意：众数可能有很多个，非一个数值

b=df_f1['单价'].mean()
print("单价均值是：\n",b) 
```

```python
f_dict={"商品类别":a[0],"单价":b}                  #填充字典

df_f1.fillna(value=f_dict,inplace=True)             #填充

print("填充后26号数据：\n" ) 
df_f1.iloc[[26]]                                  #查看第26条记录，填充结果
```

# 二、处理异常值
### (1) 删除异常值
    df.drop(index,inplace)     :删除指定索引的记录
### (2) 替换异常值
    df.loc(行、列或布尔索引)=新值

#### 例2-1 发现异常值

```python
df_2=df.copy()

pd.set_option('display.float_format', lambda x: '%.1f' % x)             #设置describe的显示格式

#根据统计信息，查看异常值情况
df_2.describe()               #可发现 单价 存在负值
```

#### 例 2-2 将单价小于0的记录删除

```python
err_index=df_2[df_2.单价<0].index                  #索引
df_2.drop(index=err_index,inplace=True)           #drop(index,inplace) 删除指定索引的记录
df_2.describe()                              #单价小于0的记录已删除
```

#### 例 2-3 将小于0的单价替换为单价平均值

```python
df_3=df.copy()
idx=df_3[ df_3['单价']<0 ].index                       #小于0的行索引

df_3.loc[ df_3['单价']<0,"单价" ]= df_3['单价'].mean(0) 
df_3.iloc[idx,]                                          #查看结果
```

# 三、处理重复值
### (1) 查找重复值
    df.duplicated(subset,keep)
        subset设置按哪些字段判断是重复记录，默认按所有字段值都相同的记录为重复记录
        keep=False：重复记录标记为True
        keep="first"：第一条重复记录标记为False
        keep="last"：最后一条重复记录标记为False
### (2) 删除重复记录
    df.drop_duplicates(subset,keep,inplace)
        keep=False：删除所有重复记录
        keep="first"：删除重复时用于保留第一条
        keep="last"：删除重复时用于保留最后一条

```python
df_3=df.copy()
```

```python
#显示重复记录
df_3[ df_3.duplicated(keep=False) ]     #所有字段相同为重复，keep=False表示重复记录标记为True，显示True的记录
```

```python
# 显示重复记录，不包括最后一条重复记录
df_3[df_3.duplicated(keep="last")]              
```

#### 例2-2 删除重复记录，保留重复记录中的第一条

```python
df_3.drop_duplicates(keep="first")
```

#  四、数据加工
### (1) 拆分字段
    一个字段分为多个字段。 例如，日期型字段拆分为年、月、日
### (2) 抽取字段
    字段抽取：抽取字段中部分信息。 例如，单位数量 字段中 单位
### (3)字段合并
    多个字段合为一个。 例如，年、月、日生成一个 日期 新字段

```python
import pandas as pd
data=pd.read_csv(r".\产品订购表_预处理.csv",encoding='ANSI')
data.info()              #注意：日期字段读入后是字符串对象，不是datetime型
```

#### 例4-1  从  生产日期 字段抽取 信息 构造 新3个字段 年 、 月 、 日

```python
data['生产日期']=pd.to_datetime(data['生产日期'])              #将生产日期字段转为日期时间型
import datetime as dt                                    #导入datatime模块，使用取年、月、日
#.dt 是 Pandas 提供的一个访问器，用于操作 datetime64 类型的列。它提供了许多有用的属性和方法
```

```python
data['年']=data['生产日期'].dt.year                      #增加“年”字段
data.head()
```

```python
data['月']=data['生产日期'].dt.month                  #增加“月”字段
data['日']=data['生产日期'].dt.day                    #增加“日”字段
data.head()
```

####  例4-2  构造新字段 “单位” ，值是“单位数量”字段的第2个字符

```python
#.str 是 Pandas 提供的一个访问器，用于对字符串类型的列进行操作，但它需要配合具体的方法或属性来完成任务。
data['单位']=data['单位数量'].str[1:2]      #先获取series的str属性，则可再在属性上使用字符串操作（包括方法）
data.head()
```

####  例4-3  删除“生产日期”字段，构造新字段 “日期” ，值来源于 年、月、日字段

```python
data['年'].dtype                          # dtype查看 年 字段的类型为int型
```

```python
data.drop("生产日期",axis=1,inplace=True)
data
```

```python
data['日期']=data['年'].astype(str)+'/'+data['月'].astype(str)+'/'+data['日'].astype(str)  #astype将数据类型转换为指定的类型
data.head()
```

```python
#将data存入 产品.csv 文件中，编码采用gbk 格式，请查看没有乱码
data.to_csv(r"./产品.csv",encoding='gbk')
```
