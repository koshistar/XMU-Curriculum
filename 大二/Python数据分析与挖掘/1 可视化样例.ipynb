# 1 折线图

```python
import numpy as np
X = np.linspace(-np.pi, np.pi, 50)           #创建等差数组
C= np.cos(X)                     #C=cos(X)
S = np.sin(X)                    #S=sin(X)

plt.plot(X,C)                                #以X为x轴数据，C为Y轴数据，画折线图
plt.plot(X,S)                               #以X为x轴数据，S为Y轴数据，画折线图

plt.show()                                  #显示图形

#可以将X = np.linspace(-np.pi, np.pi, 50)中的50改为10、5，观察图形的光滑性
```

```python
#改变颜色与线条粗细、每个点的标记
plt.plot(X, C, color="blue", linewidth=1.5, marker="o")
plt.plot(X, S, color="red",  linewidth=2.5, marker=".")
```

```python
# 设置横轴的上下限
plt.xlim(-4.0,4.0)

# 设置横轴记号
#plt.xticks(np.linspace(-4,4,9,endpoint=True))                         # numpy数组
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],                       #列表形式设置
       [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])        #格式

# 设置纵轴的上下限
plt.ylim(-1.0,1.0)

# 设置纵轴记号
plt.yticks(np.linspace(-1,1,5,endpoint=True))

plt.plot(X, C, color="blue", linewidth=1.5, marker="o")
plt.plot(X, S, color="red",  linewidth=2.5, marker=".")
```

```python
plt.plot(X, C, color="blue", linewidth=1.5, marker="o",label='cos')    #label设置图例文本
plt.plot(X, S, color="red",  linewidth=2.5, marker=".",label='sin')

plt.legend(loc='upper left')                                             #设置图例位置
```

# 3 直条图

```python
#条形图barh，left设置条形图左侧位置，默认0
n = 12
X = np.arange(n)
Y1 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)
Y2 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)

plt.barh(X, +Y1, facecolor='blue', edgecolor='white',left=0)
#plt.barh(X, -Y2, facecolor='red', edgecolor='white')      
```

```python
n = 12
X = np.arange(n)
Y1 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)
Y2 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)

plt.bar(X, +Y1, facecolor='blue', edgecolor='white')         #可以画条形
#plt.bar(X, -Y2, facecolor='red', edgecolor='white')       #可以画倒条形

for x,y in zip(X,Y1):
    plt.text(x, y, '%.2f' % y, ha='center', va= 'bottom')  #参数分别是条形上文字标记的位置、格式、水平与垂直对齐

plt.ylim(0,+1.25)
```
