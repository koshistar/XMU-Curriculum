   结果是：11    Algiers
5. 查看Year有空值的记录（注意有空值的 行索引号）
   
   行索引号是：8

   （2）将AvgTemperature的空值用x填充
   
   （3）将Year的空值用其前面的有效值填充

```python
#1. 准备数据：读入 city_temperature.csv 的数据。
import pandas as pd
df=pd.read_csv('city_temperature.csv',encoding='utf-8')
df.head()
```

```python
#2. 查看空值情况
df.info()
```

```python
#3. 删除3个全空字段
df.dropna(axis=1,how='all', inplace=True)
df
```

```python
#4. 查看AvgTemperature有空值的记录（记住：有空值的行索引号 与 City的值）
df[df['AvgTemperature'].isnull()][['City']]
```

```python
#5. 查看Year有空值的记录（注意有空值的 行索引号）
df[df['Year'].isnull()].index
```

```python
#6. （1）第4题找出了空值的City，求该City的AvgTemperature的平均值x
x=df[df['City']=='Algiers']['AvgTemperature'].mean()
x
```

```python
#6. （2）将AvgTemperature的空值用x填充
# df.fillna({'AvgTemperature': x}, inplace=True)
df['AvgTemperature'].fillna(x, inplace=True)
df
```

```python
#6. （3）将Year的空值用其前面的有效值填充
df['Year'].fillna(method='ffill', inplace=True)
# df['Year'].ffill(inplace=True)
df
```

```python
#7. 用iloc查看填充的2条记录（行索引是8与11），是否填充正确。
df.iloc[[8,11]]
```

```python
#8. 设置describe的显示格式：pd.set_option('display.float_format', lambda x: '%.1f' % x) 
    # 用describe查看统计信息，发现异常值。Day的最大值与Month的最小值异常
pd.set_option('display.float_format', lambda x: '%.1f' % x)
df.describe()
```

```python
#9. 删除Day>31、Month<1的异常记录
df=df[(df['Day']<=31)&(df['Month']>=1)]
df.describe()
```

```python
# 10. 显示重复记录
df[df.duplicated(keep=False)]
```

```python
# 11. 原地删除重复记录，保留重复记录中的最后一条
df.drop_duplicates(keep='last',inplace=True)
df
```

```python
# 12. 将处理后的数据存入 temperature_new.csv 文件中，不要索引号（index=False）
df.to_csv('temperature_new.csv', index=False)
```
