## 综合样例：
1. buy_record.xlsx文件中有商品号no、商品名name、类别category、库存storage、订购量order等信息，buy_price.xlsx文件有no、商品价格price，请将两文件中的数据按no合并，保存在df_all中
2. 筛选各类别的商品数据，并将结果（不要索引index=False）保存至各文件中，文件名为 “类别.xlsx”。例如，

    category是调料的数据，保存至“调料.xlsx”
    
    category是奶类的数据，保存至“奶类.xlsx”


```python
#读入buy_record.xlsx中的数据，保存在数据框df1中
import pandas as pd
df1=pd.read_excel(r'.\buy_record.xlsx',usecols=['no','name','category','storage','order'])
df1

```

```python
#读入buy_price.xlsx中的数据，保存在数据框df2中
df2=pd.read_excel(r'.\buy_price.xlsx',usecols=['no','price'])
df2

```

```python
#将两文件中的数据按no号合并，保存在df_all中
df_all=pd.merge(left=df1,right=df2,how='inner',on='no')
df_all

```

```python
#取出数据表中所有类别名
cate_set=set(df_all['category'])
cate_set
```

```python
#筛选各类别的商品数据，并将结果保存至各文件中，文件名为 “类别.xlsx”
for i in cate_set:
    temp=df_all[df_all['category']==i]
    temp.to_excel(i+'.xlsx')
    print(temp)
```
