```python
#china2018.csv中是china几个城市2018-2020年的气温数据

import pandas as pd
df=pd.read_csv(r'.\china2018.csv',encoding='gbk')
df.head()
```

#### 例1-1 各城市温度的总个数

```python
df['City'].value_counts()
```

#### 例1-2  统计温度的最大值、最小值、平均值

```python
df['AvgTemperature'].agg(['max', 'min', 'mean'])
```

#### 例，统计年的最大值，月的最小值，气温平均值

```python
pd.set_option('display.float_format', lambda x: '%d' % x)  #设置显示格式
df.agg({'Year':'max','Month':'min','AvgTemperature':'mean'})
```

#### 例2-1 统计各城市3年的平均气温，以城市为行索引

```python
pd.set_option('display.float_format', lambda x: '%.1f' % x)  #设置显示格式
df.groupby('City',as_index=True).agg({'AvgTemperature':'mean'})

```

#### 例2-2 统计各城市不同月的平均气温，以城市、月为行索引

```python

df.groupby(['City','Month'],as_index=True).agg({'AvgTemperature':'mean'})
```

#### 练 3-1 
    （1）统计各月各城市的平均气温，以月、城市为行索引,结果存至df4
    （2）将df4的“AvgTemperature”列标签改为 “月平均气温”
    （3）构造“等级”字段，“月平均气温”在(1,40]为low，在(40,70]为mid，在(70,100]为high.

```python
df4=df.groupby(['Month','City'],as_index=True).agg({'AvgTemperature':'mean'})
df4.columns={'月平均气温'}
df4['等级']=pd.cut(df4['月平均气温'],bins=[0,40,70,100],labels=['low','mid','high'])
df4
```

### 练3-2
    （1）构造“气温均分3等级”字段，“月平均气温”均分至3个区间.

```python
df4['气温均分3等级']=pd.cut(df4['月平均气温'],bins=3)
df4
```

#### 练习3-3  
    （1）将“日期”字段分区：按3周的时间周期，重新生成新的时间序列重采样对象.
    （2）统计各时间段内“AvgTemperature”的平均值

```python
df5=df.copy()
df5['日期']=pd.to_datetime(df5['日期'])
df6=df5.resample('3W',on='日期').agg({'AvgTemperature':'mean'})
df6

```
