#### 练习：统计story.txt中不同单词出现的次数（区分大小写），输出各单词出现的次数。
    输出样例：
    frog 12
    prince 2
    A 1
    long 3
    time 2
    ago, 1
    in 13
    a 19
    far 1
    ......

```python
fp=open("./story.txt","r")
word={}
s=fp.read()
for i in s.split():
    if i in word:
        word[i]+=1
    else:
        word[i]=1
    #word[i]=word.get(i,0)+1
for k,v in word.items():
    print(k,v)
```

#### 综合例题：统计story.txt中不同单词出现的次数（不区分大小写），
1. 输出出现次数最多的5个单词。
2. 将统计结果写入文件word_count.txt，文件中一行一个单词及其个数

```python
fp=open(r".\story.txt","r")        #以读的方式打开当前文件夹下的story.txt文件,创建指向story.txt的文件对象fp
s=fp.read()                         #read方法读fp对象中的文本，赋值给s
s=s.replace('"',' ')                #将双引号改为空格

diccount={}                                              #创建空字典
word_list=s.lower().split()       #s中的字符串全转小写，以空白字符为分隔符，拆分s（即拆分单词），保存至列表
for word in word_list:              
    diccount[word]=diccount.get(word,0)+1      #统计各单词的个数，结果存入diccount字典（{word：个数....}）
    
word_sort=sorted(diccount.keys(),key=diccount.get)   #键值 按 个数的升序排列
word_sort=word_sort[::-1]                           #改为降序
for i in word_sort[:5]:                             #遍历前5个元素（最大的5个）
    print(*(i,diccount[i]))                            #输出前5个 (单词，个数)
    

result=''                                              #初始值为空串
for i in word_sort:                                   #遍历排序后的单词列表
    result=result+i+'\t'+str(diccount[i])+'\n'          #将 统计结果 拼接成一个字符串，存至result
print(result)
fp_w=open(r".\word_count.txt","w")                #以写的方式打开word_count.txt
fp_w.write(result)                               #用write方法将字符串输出至fp_w所指的文件
fp_w.close()                                       #关闭fp_w所指的文件文件
```
