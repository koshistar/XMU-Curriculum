## 2.2、访问DataFrame对象：

### （1）. 基于索引的方式：用于访问整行或整列数据
    * 索引是**单值或者一维列表**：选择**列**。    
        
        (1) 查询某一列：data['人口']	
        
        (2) 查询多列：data[[ 0,2] ]     （注意：没有 列标签，才可用 列位置索引。）
    * 索引是**切片**：选择**行**
    
        例如，从5到14行：data[5:15]
    * 布尔型索引: 条件表达式
   
    
### （2）. loc函数：

    * DataFrame.loc[行索引名称或条件, 列索引名称]
### （3）. iloc函数：

    * DataFrame.iloc[行索引位置, 列索引位置]

### （1）  基于索引的方式访问数据框

```python
import pandas as pd
data=[ [710.1,8.4,26],[536,8.7,23],[261.1,8.1,27],[1500.1,8.6,22] ]
index=['福州','南昌','厦门', '成都']
df7=pd.DataFrame( data, index, columns=['人口','GDP增速','白天平均气温']) 
df7
```

```python
#例3-4 使用列标签索引
print("\n各城市的人口是：\n",df7["人口"])
print("\n各城市的人口、GDP增速是：\n",df7[ ["人口","GDP增速"] ])       #使用 列标签列表 指定列
```

```python
#例3-5 使用位置索引
import numpy as np
df9=pd.DataFrame(np.arange(0,12).reshape(3,4))
print("原数据框是：\n",df9)
```

```python
#位置索引访问示例，注意，没有列标签的，才可用位置索引
print("\n第1列的数据\n",df9[1])                   #列索引=1的列。因为索引是单个数值，表示选择列
print("\n第1、3列的数据\n",df9[[1,3]])            #列索引=1或3的列。因为索引是列表，表示选择列
print("\n前2行数据\n",df9[:2])                  #前2行的数据。因为索引是切片，表示选择行
print("\n前2行第2列的数据\n",df9[:2][2])         #前2行第2列的数据。df9[:2]选择前2行后，再选择第2列
print("\n第3行第3列的数据\n",df9[2][2])      #行索引=2 and 列索引=2的数据。df9[2]选择第2列后，成为一维，再选择第2行
```

### （2） 使用loc、iloc访问数据框

```python
#复制一份df7的副本
df10=df7.copy()
df10
```

```python
#loc函数示例
print("\n南昌的人口是：\n",df10.loc["南昌","人口"])
print("\n南昌-成都的人口是：\n",df10.loc["南昌":"成都","人口"])        #使用行标签切片，左闭右闭
print("\n南昌-成都的人口、GDP增速是：\n",df10.loc["南昌":"成都",["人口","GDP增速"]])   #使用 列标签列表 指定列
print("\n所有城市的人口、GDP增速是：\n",df10.loc[:,["人口","GDP增速"]])   #使用：表示所有
```

```python
#iloc函数示例
print("\n南昌的人口是：\n",df10.iloc[1,0])
print("\n南昌-成都的人口是：\n",df10.iloc[1:,0])                #使用  行索引位置切片
print("\n南昌-成都的人口、GDP增速是：\n",df10.iloc[1:,[1,0]])   #使用 列索引位置列表 指定列
print("\n所有城市的人口、GDP增速是：\n",df10.iloc[:,[1,0]])     #使用：表示所有
```

#### 练习7：使用loc查询df10中
1. 福州、厦门的人口
2. 福州至厦门所有行的记录
3. 所有城市的人口与白天平均气温

```python
print("\n福州、厦门的人口是:\n",df10.loc[["福州","厦门"],"人口"])
print("\n福州至厦门所有行的记录是:\n",df10.loc["福州":"厦门",:])
print("\n所有城市的人口与白天的平均气温\n",df10.loc[:,["人口","白天平均气温"]])
```

#### 练习8：使用iloc查询df10中
1. 福州、厦门的人口
2. 福州至厦门所有行的记录
3. 所有城市的人口与白天平均气温

```python
print("\n福州、厦门的人口是:\n",df10.iloc[[0,2],0])
print("\n福州至厦门所有行的记录是:\n",df10.iloc[0:3,:])
print("\n所有城市的人口与白天的平均气温\n",df10.iloc[:,[0,2]])
```

### 3   按条件查询：

(1) 使用布尔型索引访问：
* DataFrame[布尔值条件]       
* DataFrame.loc[布尔型条件, 列索引]

(2) pandas条件常用的方法
* between(i,j)：判断是否在[i,j]内
* isin(list)  ：判断是否列表中的元素
* isnull    ： 空值NaN
* &        ：并且  运算符   
        注意：多个条件组合查询时，每个条件必须写在()中
    
        例如， (条件1) & (条件2)
* |        ：或者  运算符
* ~        ：非   运算符


```python
#例4-1 查询df10中人口超过600万的城市
print(df10)
df10[ df10["人口"]>600 ]
```

```python
#例4-2 查询 白天平均气温在[25,28]之间的城市
df10[ df10['白天平均气温'].between(25,27) ]
```

```python
#例4-3 查询 白天平均气温 是23、27的记录，只显示人口、白天平均气温 列
#df10[ df10["白天平均气温"].isin([23,27]) ][["人口","白天平均气温"]]
df10.loc[df10["白天平均气温"].isin([23,27]),["人口","白天平均气温"]]
```

```python
# 例4-4 查询 GDP增速 >8.5 并且 人口 >800 的记录
df10[ (df10["GDP增速"]>8.5)  & (df10['人口']>800)  ]    #注意，& 的每个条件写在()中
```

#### 练习9：练习条件查询，用loc方法访问df10，完成上面4个例题（例4-1至例4-4）

```python
print(df10.loc[df10["人口"]>600])
print(df10.loc[df10["白天平均气温"].between(25,27)])
print(df10.loc[df10["白天平均气温"].isin([23,27]),["人口","白天平均气温"]])
print(df10.loc[(df10["GDP增速"]>8.5)&(df10["人口"]>800)])
```
